```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
ax = plt.subplot(221)

plt.plot([.5, .25, 0], color=sns.color_palette()[0], marker='o')
plt.fill_between([0, 1, 2.], [.4, 0.15, -.1], [.6, .35, .1], color=sns.color_palette()[0], alpha=0.3)

# plt.title("Willingness to pay")
plt.ylabel('Willingness to pay')

# plt.title("Small probability (15%)")
plt.axhline(0.0, ls='--', color='k')

plt.annotate('Risk averse', xy=(0.5, 0.05), xytext=(0, 0.05),
             textcoords='offset points', ha='center')

plt.annotate('Risk seeking', xy=(0.5, -0.15), xytext=(0, 0.0),
             textcoords='offset points', ha='center')

plt.xlabel('Stake size')

plt.subplot(222, sharex=ax, sharey=ax)

plt.plot([1, .5, 0], marker='o')

plt.fill_between([0, 1, 2.], [.75, 0.25, -.25], [1.25, .75, .25], color=sns.color_palette()[1], alpha=0.3)

plt.ylim(-.2, 1.1)

plt.plot([1, .5, 0], marker='o')

plt.fill_between([0, 1, 2.], [.75, 0.25, -.25], [1.25, .75, .25], color=sns.color_palette()[1], alpha=0.3)


plt.axhline(0.0, ls='--', color='k')
plt.xticks([])
plt.yticks([])

plt.xlabel('Stake size')

plt.annotate('Risk averse', xy=(0.5, 0.05), xytext=(0, 0.05),
             textcoords='offset points', ha='center')

plt.annotate('Risk seeking', xy=(0.5, -0.15), xytext=(0, 0.0),
             textcoords='offset points', ha='center')

# plt.title("Larger probability (55%)")
sns.despine()

from scipy.stats import norm

# Get base colors from sns.color_palette
base_colors = sns.color_palette()

# Function to adjust saturation
def adjust_saturation(color, factor):
    return sns.set_hls_values(color, l=None, s=factor)

colors = [adjust_saturation(base_colors[0], 0.3 + 0.15 * i) for i in range(5)]
colors += [adjust_saturation(base_colors[1], 0.3 + 0.15 * i) for i in range(5)]

ax = plt.subplot(223)

x = np.linspace(0, 1, 100)

for mean, c in zip(np.linspace(.25, .75, 5), colors):
    y = norm(mean, 0.1).pdf(x)

    plt.fill_between(x, y, alpha=0.3, color=c)

sns.despine()
plt.xlabel('Payoff magnitude')

plt.ylabel('Decoded posterior density')
plt.xticks([])
plt.yticks([])

plt.subplot(224, sharex=ax, sharey=ax)
colors = [adjust_saturation(base_colors[1], 0.3 + 0.15 * i) for i in range(5)]
colors += [adjust_saturation(base_colors[2], 0.3 + 0.15 * i) for i in range(5)]


for mean, c in zip(np.linspace(.25, .75, 5), colors):
    y = norm(mean, 0.2).pdf(x)

    plt.fill_between(x, y, alpha=0.3, color=c, )

sns.despine()
plt.xlabel('Payoff magnitude')

```
